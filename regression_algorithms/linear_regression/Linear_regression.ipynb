# Thuật toán Linear Regression
Linear Regression là một thuật toán machine learning cơ bản và đơn giản nhất. Cơ chế của thuật toán dựa trên công thức hồi quy tuyến tính có dạng phương trình: y = ax + b.
Khi có tập dữ liệu có X và y, sau nhiều lần train ta thu được hệ số a và b, từ đó suy ra dạng phương trình cần mô hình.
Hạn chế của Linear Regression là rất nhạy cảm với nhiễu.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
# Make data:
X = np.random.rand(100, 1)
y = 4 + 3*X + np.random.rand(100, 1)
```

```python
# Visualize data:
plt.scatter(X, y)
plt.xlabel('X values')
plt.ylabel('y values')
plt.show()
```

```python
# Training data:
lin_reg.fit(X, y)
```

```python
# y = ax + b
a, b = lin_reg.intercept_, lin_reg.coef_
print(a, b)
```

```python
# Visualize data and line:
X_new = np.array([[0], [2]])
y_predict = lin_reg.predict(X_new)

plt.plot(X_new, y_predict, "r-", label='Predict')
plt.scatter(X, y)
plt.axis([0, 1.25, 0, 10])
plt.show()
```
